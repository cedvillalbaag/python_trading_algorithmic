## Calculate Correlation & Covariance - Stocks

La correlación estadística es una medida que indica la relación entre dos o más variables. Es una herramienta que se utiliza para describir relaciones simples y evaluar la fuerza y la dirección de esa relación. 

**Tipos de correlación** 

- Correlación positiva: Las variables aumentan o disminuyen juntas.
- Correlación negativa: Una variable aumenta mientras la otra disminuye.
- Correlación débil: El coeficiente de correlación está cerca de cero.
- Correlación fuerte: El coeficiente de correlación está cerca de 1 o -1.

**Coeficientes de correlación**
- El coeficiente de Pearson es el más comúnmente utilizado para medir la relación lineal entre dos variables continuas. 
- El coeficiente de Spearman mide la relación entre las posiciones relativas de los datos. 
- El coeficiente de Kendall mide la relación ordinal entre dos variables. 

**Aplicaciones**
El análisis de correlación se aplica en diversos campos, como la economía, la psicología, la ingeniería, la física, la epidemiología o en inteligencia de negocios. 


**Covariance**
La covarianza es una medida estadística que indica la relación entre dos variables aleatorias. Es decir, la covarianza muestra cómo se mueven dos variables en conjunto.

**¿Qué se puede hacer con la covarianza?**
- Se utiliza en finanzas para determinar qué valores de activos incluir en una cartera. 
- Se utiliza en economía, ingeniería y machine learning. 
- Se utiliza para reducir el riesgo y la volatilidad en una cartera. 
- Se utiliza para encontrar una base óptima para representar los datos. 

**¿Cómo se interpreta la covarianza?**
- El signo de la covarianza indica la dirección de la relación entre las variables.
- Un signo positivo indica que las variables tienden a moverse en la misma dirección.
- Un signo negativo indica que las variables tienden a moverse en direcciones opuestas.

**¿Cómo se calcula la covarianza?**
- La covarianza entre dos variables aleatorias X y Y se puede denotar como Cov(X, Y). 
- La covarianza se mide en unidades, que se calculan multiplicando las unidades de las dos variables.

## Forward Fill | Backward Fill Methods
En trading de stocks, los métodos forward fill (ffill) y backfill (bfill) son técnicas de imputación de datos utilizadas para manejar valores faltantes en series temporales, especialmente en precios de acciones, volúmenes y otros indicadores financieros.

### Forward Fill (ffill) → Relleno hacia adelante
🔹 Llena los valores faltantes con el último valor conocido.

Es útil cuando queremos asumir que el último dato válido sigue siendo válido hasta que haya un nuevo valor.
Se usa comúnmente en datos de precios cuando hay huecos en la información (ejemplo: días sin trading, problemas en la fuente de datos).

### Backfill (bfill) → Relleno hacia atrás
🔹 Llena los valores faltantes con el siguiente valor disponible.

🔹 Se usa cuando queremos que los valores faltantes tomen el siguiente dato válido disponible.

## Cumulative Returns & Drawdowns Metrics

***1. Cumulative Returns (Retornos Acumulados)***   
Los retornos acumulados muestran cómo ha crecido o decrecido una inversión a lo largo del tiempo, considerando la reinversión de ganancias. Se calcula multiplicando los rendimientos diarios de forma sucesiva.

📌 Interpretación:

- Si el retorno acumulado es positivo, la inversión ha crecido.  
- Si es negativo, la inversión ha perdido valor

---

***2. Drawdowns (Caídas o Pérdidas desde un Máximo)***   
El drawdown mide la disminución de una inversión desde su punto más alto hasta su punto más bajo antes de recuperarse. Es una medida clave de riesgo.

📌 Interpretación:

- Si el drawdown es alto, significa que la inversión sufrió pérdidas significativas en algún momento.  
- Se usa para calcular el Max Drawdown, que es la caída más grande en un período determinado

---

🔥 **¿Cómo se usan en trading?**  
✅ Cumulative Returns se usa para evaluar el crecimiento total de la inversión y comparar diferentes estrategias.  
✅ Drawdown ayuda a medir el riesgo y la volatilidad de una estrategia. Un drawdown alto indica una estrategia riesgosa.

📉 Ejemplo real:
Si dos estrategias tienen el mismo retorno acumulado, la que tenga menor drawdown es preferible porque tuvo menos caídas pronunciadas.

### 📈 SMA Strategy en Trading (Estrategia con Media Móvil Simple)

Una **SMA Strategy** (Estrategia con **Simple Moving Average**, o **Media Móvil Simple - SMA**) es una estrategia de trading basada en el uso de medias móviles para identificar tendencias y generar señales de compra o venta.

---

#### 🔹 ¿Qué es la SMA (Simple Moving Average)?
La **SMA** es un indicador técnico que calcula el **promedio de precios** de un activo en un período determinado. Suaviza la volatilidad y ayuda a identificar la dirección de la tendencia.

#### 📌 Fórmula de SMA:
\[
SMA_t = \frac{P_1 + P_2 + ... + P_n}{n}
\]
Donde:
- \( P_1, P_2, ..., P_n \) son los precios de cierre en los últimos \( n \) períodos.
- \( n \) es el número de períodos de la media móvil.

Ejemplo:
- **SMA 50** → Promedio de los últimos 50 días.
- **SMA 200** → Promedio de los últimos 200 días (muy usada para tendencias de largo plazo).

---

### 🔹 Tipos de estrategias SMA en Trading

#### ✅ 1. Cruce de Medias Móviles (Golden & Death Cross)
Se basa en la interacción entre dos SMA de diferentes períodos:
- **Golden Cross** (Cruz Dorada) → SMA **rápida** cruza hacia arriba la SMA **lenta** → **Señal de compra** ✅
- **Death Cross** (Cruz de la Muerte) → SMA **rápida** cruza hacia abajo la SMA **lenta** → **Señal de venta** ❌

📌 **Ejemplo de configuración:**
- **SMA 50** (rápida)
- **SMA 200** (lenta)

---

#### ✅ 2. SMA como Soporte y Resistencia Dinámica
Los traders usan la SMA para encontrar **niveles clave**:
- Si el precio está **por encima de la SMA**, indica tendencia alcista y la SMA actúa como **soporte**.
- Si el precio está **por debajo de la SMA**, indica tendencia bajista y la SMA actúa como **resistencia**.

📌 **Ejemplo:**
- La **SMA 200** se usa en gráficos diarios como referencia de **soporte/resistencia** para tendencias de largo plazo.

---

## Creating our First Class

````python
class SMABacktester():
    def __init__(self, symbol, SMA_S, SMA_L, start, end):
        self.symbol = symbol
        self.SMA_S = SMA_S
        self.SMA_L = SMA_L
        self.start = start
        self.end = end
        self.results = None
        self.get_data() 
        
    def get_data(self):
        df = yf.download(self.symbol, start=self.start, end=self.end)
        df = df.rename(columns={self.symbol: 'Close'})
        data = df.Close
        data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
        data["SMA_S"] = data.Close.rolling(self.SMA_S).mean()
        data["SMA_L"] = data.Close.rolling(self.SMA_L).mean()
        data.dropna(inplace=True)
        self.data2 = data
        
        return data
    
    def test_results(self):
        data = self.data2.copy().dropna()
        data["position"] = np.where(data["SMA_S"]>data["SMA_L"],1,-1)
        data["strategy"] = data["returns"]*data.position.shift(1)
        data.dropna(inplace=True)   
        data["returnsbh"] = data["returns"].cumsum().apply(np.exp)
        data["returnsstrategy"] = data["strategy"].cumsum().apply(np.exp)        
        perf = data["returnsstrategy"].iloc[-1]
        outperf = perf - data["returnsbh"].iloc[-1]
        self.results = data
        
        ret = np.exp(data["strategy"].sum())
        std = data["strategy"].std()*np.sqrt(252)
        
        #return ret, std
        return round(perf,6), round(outperf,6)
    
    def plot_results(self):
        if self.results is None: 
            print("run the test please")
        else:
            title= "{}| SMA_S={} | SMA_L{}".format(self.symbol, self.SMA_S, self.SMA_L)
            self.results[["returnsbh","returnsstrategy"]].plot(title=title, figsize = (12,8))
````
